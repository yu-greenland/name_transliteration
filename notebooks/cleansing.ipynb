This note book will be dedicated to cleansing the Twitter data.

This is part of the cleansing stage described in the Design Document.

## input:
    - filtered Twitter user name data in the form of csv
## output:
    - a csv file with similar user name and screen name
    
    - hopefully a good dataset to train on

```python
import pandas as pd
ja_data = pd.read_csv('ja_data.csv')
ja_data.dropna(inplace=True)
ja_data.reset_index(drop=True, inplace=True)

ja_data.head()
```

```python
import pykakasi
import editdistance

edit_threshold = 4
rows_over_threshold = []
ja_data_small = ja_data.iloc[:3000]
for index, row in ja_data_small.iterrows():
    username = row.username
    screen_name = row.screen_name
    kks = pykakasi.kakasi()
    translit = kks.convert(username)
    roman = ''
    for item in translit:
        roman = roman + item['hepburn']
    edit_distance = editdistance.eval(roman, screen_name)
    if edit_distance > edit_threshold:
        rows_over_threshold.append(index)
```

```python
ja_data_small.drop(rows_over_threshold)
```
